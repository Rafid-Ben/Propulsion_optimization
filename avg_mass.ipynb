```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from matplotlib.patches import FancyArrowPatch
from matplotlib.ticker import PercentFormatter
```

```python
m0=100
deltav=0.8
c=0.4
n=5
mp= m0*(1-np.exp(-deltav/c))
mpi= m0*(1-np.exp(-(1/n)*deltav/c))

mf=m0-mp
mfi=m0-mpi
deltavi=deltav/n

tau=n*mpi/mp
ti=tau/n

```

```python

# Define a combination of Gaussian functions with non-symmetric minima and maxima
def gaussian_surface(x, y):
    g1 = -2 * np.exp(-((x - 2)**2 / 0.5 + (y - 2)**2 / 0.8))  # Gaussian 1 (peak at 4)
    g2 = -1.5 * np.exp(-((x + 2)**2 / 0.6 + (y + 2)**2 / 0.4))  # Gaussian 2 (valley at -4)
    g3 = -1 * np.exp(-((x - 1)**2 / 0.7 + (y + 1)**2 / 0.9))  # Gaussian 3 (peak at 3)
    g4 = -0.5 * np.exp(-((x + 1)**2 / 0.3 + (y - 1)**2 / 0.6))  # Gaussian 4 (valley at -3)
    return g1 + g2 + g3 + g4  # Combine Gaussians


# Create a grid of x and y values
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
x, y = np.meshgrid(x, y)

# Calculate the function values
z = gaussian_surface(x, y)

# Create a 3D plot
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface

surf = ax.plot_surface(x, y, z, cmap='plasma', edgecolor='k', linewidth=0.4, alpha=0.8)


# Customize the view
#ax.set_title('A Non-Convex Combination of Gaussian Distributions')

ax.view_init(elev=15, azim=115)   # Adjust view angle
#fig.colorbar(surf, shrink=0.5, aspect=10, label='f(x, y)')
ax.xaxis.set_tick_params(labelbottom=False)
ax.yaxis.set_tick_params(labelleft=False)
ax.zaxis.set_tick_params(labelleft=False)

plt.savefig("figures/electric/minima.eps", bbox_inches="tight")
# Show the plot
plt.show()
```
